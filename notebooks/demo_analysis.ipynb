## 1. Setup and Imports

## 2. Initialize Pipeline

Create pipeline instance pointing to our data file.

```python
pipeline = AirQualityPipeline('../data/openaq_sample.parquet')
print("Pipeline initialized")
```

## 3. Run Complete Analysis

This executes the full pipeline:
- Load data with column projection
- Clean and preprocess
- Compute statistics
- Generate visualizations

```python
results = pipeline.run_analysis()
```

## 4. Explore Results

### 4.1 Cleaned Data

```python
df_cleaned = results['cleaned_data']

print(f"Shape: {df_cleaned.shape}")
print(f"\nColumns: {df_cleaned.columns.tolist()}")
print(f"\nFirst 5 rows:")
df_cleaned.head()
```

### 4.2 Top 10 Most Polluted Cities

```python
avg_pm25_by_city = results['stats']['avg_pm25_by_city']

print("Top 10 cities by PM2.5 concentration:")
print("=" * 50)
for i, (city, pm25) in enumerate(avg_pm25_by_city.head(10).items(), 1):
    print(f"{i:2d}. {city:30s} {pm25:8.2f} µg/m³")
```

### 4.3 Monthly NO₂ Trends

```python
avg_no2_by_month = results['stats']['avg_no2_by_month']

print(f"Months analyzed: {len(avg_no2_by_month)}")
print(f"Average NO₂: {avg_no2_by_month.mean():.2f} µg/m³")
print(f"\nFirst 10 months:")
print(avg_no2_by_month.head(10))
```

### 4.4 Data Quality by Country

```python
missing_pm25 = results['stats']['missing_pm25_per_country']

print("Top 10 countries by % missing PM2.5:")
print("=" * 50)
for i, (country, pct) in enumerate(missing_pm25.head(10).items(), 1):
    print(f"{i:2d}. {country:30s} {pct:6.2f}%")
```

## 5. Statistical Summary

```python
print("ANALYSIS SUMMARY")
print("=" * 60)
print(f"Total observations:      {len(df_cleaned):,}")
print(f"Cities analyzed:         {len(avg_pm25_by_city):,}")
print(f"Countries analyzed:      {len(missing_pm25):,}")
print(f"Months with NO₂ data:    {len(avg_no2_by_month):,}")
print(f"\nMost polluted city:      {avg_pm25_by_city.index[0]}")
print(f"Highest PM2.5:           {avg_pm25_by_city.iloc[0]:.2f} µg/m³")
print(f"\nAverage NO₂ (overall):   {avg_no2_by_month.mean():.2f} µg/m³")
print("=" * 60)
```

## 6. Key Insights

From the analysis, we can conclude:

1. **Geographic Pollution Hotspots**: Mexican cities (Poza Rica, Guadalupe, Durango) show extremely high PM2.5 levels, exceeding WHO guidelines by 40-100x

2. **Temporal Patterns**: NO₂ levels show monthly variation over 97 months (2016-2024), useful for trend analysis

3. **Data Quality Issues**: European countries show significant PM2.5 data gaps (100% missing in San Marino, Oman, Denmark), likely due to different monitoring priorities

4. **Pollutant Correlation**: Strong PM2.5-NO₂ correlation in Belgrade (R²=0.432) suggests common emission sources (likely traffic and industry)
